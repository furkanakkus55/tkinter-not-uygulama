```python
import tkinter as tk
from tkinter import messagebox

sifre = "5555"

# Şifre doğruysa not defteri penceresini açan fonksiyonu yazdım
def not_defteri_ac():
    def not_kaydet():
        yeni_not = not_alani.get("1.0", tk.END).strip()
        if yeni_not:
            with open("not.txt", "a", encoding="utf-8") as f:
                f.write(yeni_not + "\n" + "-" * 40 + "\n")
            son_not_label.config(text="Son Not:\n" + yeni_not)
            not_alani.delete("1.0", tk.END)
        else:
            messagebox.showwarning("Uyarı", "Boş not kaydedilemez.")

    # Çıkış yapılırken tüm notları gösterip pencereyi kapatacak fonksiyonu yazdım
    def cikis_yap():
        try:
            with open("not.txt", "r", encoding="utf-8") as f:
                notlar = f.read().strip()
                messagebox.showinfo("Tüm Notlar", "Kaydedilen Notlar:\n\n" + notlar)
        except FileNotFoundError:
            messagebox.showinfo("Bilgi", "Henüz hiç not kaydedilmemiş.")
        not_pencere.destroy()  # Not penceresini kapattım
        pencere.destroy()      # Ana pencereyi tamamen kapattım

    not_pencere = tk.Toplevel()
    not_pencere.title("Not Defteri")

    son_not_label = tk.Label(not_pencere, text="Son Not:\nHenüz not girilmedi.", justify="left")
    son_not_label.pack(pady=5)

    not_alani = tk.Text(not_pencere, width=50, height=10)
    not_alani.pack(pady=5)

    kaydet_btn = tk.Button(not_pencere, text="Kaydet", command=not_kaydet)
    kaydet_btn.pack(pady=5)

    # Çıkış butonunu ekledim
    cikis_btn = tk.Button(not_pencere, text="Çıkış", command=cikis_yap)
    cikis_btn.pack(pady=10)

    try:
        with open("not.txt", "r", encoding="utf-8") as f:
            tum_notlar = f.read().strip().split("-" * 40)
            if tum_notlar and tum_notlar[-1].strip():
                son_not_label.config(text="Son Not:\n" + tum_notlar[-1].strip())
    except FileNotFoundError:
        pass

# Şifreyi kontrol ettiğim fonksiyon
def kontrol():
    if giris.get() == sifre:
        messagebox.showinfo("Başarılı", "Şifre doğru. Not defteri açılıyor...")
        pencere.withdraw()
        not_defteri_ac()
    else:
        messagebox.showerror("Hata", "Şifre yanlış")

# Ana pencereyi ve giriş arayüzünü oluşturdum
pencere = tk.Tk()
pencere.title("Şifreli Not Girişi")

etiket = tk.Label(pencere, text="Şifre girin:")
etiket.pack(pady=5)

giris = tk.Entry(pencere, show="*")
giris.pack(pady=5)

btn = tk.Button(pencere, text="Giriş Yap", command=kontrol)
btn.pack(pady=5)

pencere.mainloop()


```
